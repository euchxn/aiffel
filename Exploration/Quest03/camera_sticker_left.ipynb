```python
import os 
import cv2
import matplotlib.pyplot as plt 
import numpy as np 
import dlib 
print("ðŸŒ«ðŸ›¸")
```

```python
home_dir = os.getenv('HOME')
my_image_path = os.path.join(home_dir, 'work/camera_sticker/images/myface_left.png') # ë³¸ì¸ ì´ë¯¸ì§€ê°€ ìžˆëŠ” ê²½ë¡œë¥¼ ê°€ì§€ê³  ì˜¨ë‹¤.
img_bgr = cv2.imread(my_image_path)    # OpenCVë¡œ ì´ë¯¸ì§€ë¥¼ ë¶ˆëŸ¬ì˜µë‹ˆë‹¤
img_show = img_bgr.copy()      # ì¶œë ¥ìš© ì´ë¯¸ì§€ë¥¼ ë”°ë¡œ ë³´ê´€í•©ë‹ˆë‹¤
plt.imshow(img_bgr) # ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•˜ê¸° ìœ„í•´ ì¶œë ¥í•  ì´ë¯¸ì§€ë¥¼ ì˜¬ë ¤ì¤€ë‹¤. (ì‹¤ì œ ì¶œë ¥ì€ í•˜ì§€ ì•ŠìŒ)
plt.show() # ì´ë¯¸ì§€ë¥¼ ì¶œë ¥í•´ì¤€ë‹¤. (ì‹¤ì œ ì¶œë ¥)
```

```python
# plt.imshow ì´ì „ì— RGB ì´ë¯¸ì§€ë¡œ ë°”ê¾¸ëŠ” ê²ƒì„ ìžŠì§€ë§ˆì„¸ìš”.
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb) # ì´ë¯¸ì§€ ì‹¤ì œ ì¶œë ¥í•˜ê¸° ìœ„í•´ ì´ë¯¸ì§€ë¥¼ ì˜¬ë ¤ì¤Œ (ì›í•˜ëŠ” ì‚¬ì´ì¦ˆì˜ í”½ì…€ì„ ì›í•˜ëŠ” ìƒ‰ìœ¼ë¡œ ì±„ì›Œì„œ ë§Œë“  ê·¸ë¦¼ì´ë¼ê³  í•©ë‹ˆë‹¤.)
plt.show() # ì´ë¯¸ì§€ ì‹¤ì œ ì¶œë ¥
```

```python
# detectorë¥¼ ì„ ì–¸í•©ë‹ˆë‹¤
detector_hog = dlib.get_frontal_face_detector() # ê¸°ë³¸ ì–¼êµ´ ê°ì§€ê¸°ë¥¼ ë°˜í™˜
print("ðŸŒ«ðŸ›¸")
```

```python
img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
dlib_rects = detector_hog(img_rgb, 1)   # (image, num of image pyramid)
print("ðŸŒ«ðŸ›¸")
```

```python
# ì°¾ì€ ì–¼êµ´ ì˜ì—­ ë°•ìŠ¤ ë¦¬ìŠ¤íŠ¸
# ì—¬ëŸ¬ ì–¼êµ´ì´ ìžˆì„ ìˆ˜ ìžˆìŠµë‹ˆë‹¤
print(dlib_rects)

for dlib_rect in dlib_rects: # ì°¾ì€ ì–¼êµ´ ì˜ì—­ì˜ ì¢Œí‘œ
    l = dlib_rect.left() # ì™¼ìª½
    t = dlib_rect.top() # ìœ„ìª½
    r = dlib_rect.right() # ì˜¤ë¥¸ìª½
    b = dlib_rect.bottom() # ì•„ëž˜ìª½

    cv2.rectangle(img_show, (l,t), (r,b), (0,255,0), 2, lineType=cv2.LINE_AA) # ì‹œìž‘ì ì˜ ì¢Œí‘œì™€ ì¢…ë£Œì  ì¢Œí‘œë¡œ ì§ê° ì‚¬ê°í˜•ì„ ê·¸ë¦¼

img_show_rgb =  cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
plt.imshow(img_show_rgb)
plt.show()
```

```python
model_path = os.path.join(home_dir, 'work/camera_sticker/models/shape_predictor_68_face_landmarks.dat')
    # ì €ìž¥í•œ landmark ëª¨ë¸ì˜ ì£¼ì†Œë¥¼ model_path ë³€ìˆ˜ì— ì €ìž¥
landmark_predictor = dlib.shape_predictor(model_path)
    # dlib ë¼ì´ë¸ŒëŸ¬ë¦¬ì˜ shape_predictor í•¨ìˆ˜ë¥¼ ì´ìš©í•˜ì—¬ ëª¨ë¸ì„ ë¶ˆëŸ¬ì˜´
    # landmark_predictorëŠ” RGBì´ë¯¸ì§€ì™€ dlib.rectangleë¥¼ ìž…ë ¥ ë°›ê³  dlib.full_object_detectionë¥¼ ë°˜í™˜
    # dlib.rectangle: ë‚´ë¶€ë¥¼ ì˜ˆì¸¡í•˜ëŠ” ë°•ìŠ¤
    # dlib.full_object_detection: ê° êµ¬ì„± ìš”ì†Œì˜ ìœ„ì¹˜ì™€, ì´ë¯¸ì§€ ìƒì˜ ê°ì²´ì˜ ìœ„ì¹˜ë¥¼ ë‚˜íƒ€ëƒ„
print("ðŸŒ«ðŸ›¸")
```

```python
list_landmarks = []
    # ëžœë“œë§ˆí¬ì˜ ìœ„ì¹˜ë¥¼ ì €ìž¥í•  list ìƒì„±

# ì–¼êµ´ ì˜ì—­ ë°•ìŠ¤ ë§ˆë‹¤ face landmarkë¥¼ ì°¾ì•„ëƒ…ë‹ˆë‹¤
# face landmark ì¢Œí‘œë¥¼ ì €ìž¥í•´ë‘¡ë‹ˆë‹¤
for dlib_rect in dlib_rects:
    points = landmark_predictor(img_show, dlib_rect)
        # ëª¨ë“  landmarkì˜ ìœ„ì¹˜ì •ë³´ë¥¼ points ë³€ìˆ˜ì— ì €ìž¥
    list_points = list(map(lambda p: (p.x, p.y), points.parts()))
        # ê°ê°ì˜ landmark ìœ„ì¹˜ì •ë³´ë¥¼ (x,y) í˜•íƒœë¡œ ë³€í™˜í•˜ì—¬ list_points ë¦¬ìŠ¤íŠ¸ë¡œ ì €ìž¥
    list_landmarks.append(list_points)
        # list_landmarksì— ëžœë“œë§ˆí¬ ë¦¬ìŠ¤íŠ¸ë¥¼ ì €ìž¥

print(len(list_landmarks[0]))
    # ì–¼êµ´ì´ nê°œì¸ ê²½ìš° list_landmarksëŠ” nê°œì˜ ì›ì†Œë¥¼ ê°–ê³ 
    # ê° ì›ì†ŒëŠ” 68ê°œì˜ ëžœë“œë§ˆí¬ ìœ„ì¹˜ê°€ ë‚˜ì—´ëœ list
    # list_landmarksì˜ ì›ì†Œê°€ 1ê°œì´ë¯€ë¡œ list_landmarks[1]ì„ í˜¸ì¶œí•˜ë©´ IndexErrorê°€ ë°œìƒ
```

```python
# list_landmarksì˜ ì›ì†Œê°€ 1ê°œì´ë¯€ë¡œ ì•„ëž˜ ë°˜ë³µë¬¸ì€ í•œë²ˆë§Œ ì‹¤í–‰ë¨
for landmark in list_landmarks:
    for point in landmark:
        cv2.circle(img_show, point, 2, (0, 255, 255), -1)
            # cv2.circle: OpenCVì˜ ì›ì„ ê·¸ë¦¬ëŠ” í•¨ìˆ˜
            # img_show ì´ë¯¸ì§€ ìœ„ ê°ê°ì˜ pointì—
            # í¬ê¸°ê°€ 2ì´ê³  (0, 255, 255)ìƒ‰ìœ¼ë¡œ ë‚´ë¶€ê°€ ì±„ì›Œì§„(-1) ì›ì„ ê·¸ë¦¼
            # (ë§ˆì§€ë§‰ ì¸ìˆ˜ê°€ ìžì—°ìˆ˜ë¼ë©´ ê·¸ë§Œí¼ì˜ ë‘ê»˜ì˜ ì„ ìœ¼ë¡œ ì›ì´ ê·¸ë ¤ì§)

img_show_rgb = cv2.cvtColor(img_show, cv2.COLOR_BGR2RGB)
    # RGB ì´ë¯¸ì§€ë¡œ ì „í™˜
plt.imshow(img_show_rgb)
    # ì´ë¯¸ì§€ë¥¼ ì¤€ë¹„
plt.show()
    # ì´ë¯¸ì§€ë¥¼ ì¶œë ¥
```

```python
for dlib_rect, landmark in zip(dlib_rects, list_landmarks): # ì–¼êµ´ ì˜ì—­ì„ ì €ìž¥í•˜ê³  ìžˆëŠ” ê°’ê³¼ 68ê°œì˜ ëžœë“œë§ˆí¬ë¥¼ ì €ìž¥í•˜ê³  ìžˆëŠ” ê°’ìœ¼ë¡œ ë°˜ë³µë¬¸ ì‹¤í–‰
    print (landmark[30]) # ì½”ì˜ indexëŠ” 30 ìž…ë‹ˆë‹¤
    x = landmark[33][0] # ì´ë¯¸ì§€ì—ì„œ ì½” ë¶€ìœ„ì˜ xê°’
    y = landmark[33][1] - dlib_rect.height()//2 # ì´ë¯¸ì§€ì—ì„œ ì½” ë¶€ìœ„ì˜ yê°’ - ì–¼êµ´ ì˜ì—­ì˜ ì„¸ë¡œë¥¼ ì°¨ì§€í•˜ëŠ” í”½ì…€ì˜ ìˆ˜//2 â†’ (437, 182-(186+1//2))
    w = h = dlib_rect.width() # ì–¼êµ´ ì˜ì—­ì˜ ê°€ë¡œë¥¼ ì°¨ì§€í•˜ëŠ” í”½ì…€ì˜ ìˆ˜ (531-345+1) â†’ max(x) - min(x) +1(í”½ì…€ì˜ ìˆ˜ ì´ê¸° ë•Œë¬¸ì— 1ì„ ë”í•´ì¤Œ â†’ í”½ì…€ ìˆ˜ëŠ” ì  í•˜ë‚˜ë¡œë„ 1ì´ ë¨)
    print (f'(x,y) : ({x},{y})')
    print (f'(w,h) : ({w},{h})')
```

```python
sticker_path = os.path.join(home_dir, 'work/camera_sticker/images/cat-whiskers.png') # ì´ë¯¸ì§€ì˜ ê²½ë¡œ
img_sticker = cv2.imread(sticker_path)
img_sticker = cv2.resize(img_sticker, (w,h))
print (img_sticker.shape)
```

```python
refined_x = x - w // 2 # 437 - (187//2) = 437-93 = 344
refined_y = y #- h # 89-187 = -98

print (f'(x,y) : ({refined_x},{refined_y})') # ìŒìˆ˜ ë°œìƒ : ì´ë¯¸ì§€ ë²”ìœ„ë¥¼ ë²—ì–´ë‚¨.
```

```python
sticker_area = img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]]

img_show[refined_y:refined_y+img_sticker.shape[0], refined_x:refined_x+img_sticker.shape[1]] = \
    np.where(img_sticker==255,sticker_area,img_sticker).astype(np.uint8)
print("ìŠ~")
```
